

**Inverting an Image**


The task of our first filter is to invert any 8\-bit grayscale image to turn a positive image into a negative image. Inverting the intensity of an image is a typical point operation.


8\-bit grayscale images have pixel values ranging from 0 (black) to 255 (white), and we assume that the width and height of the image are M and N, respectively.


The operation is very simple: the value of each image pixel *I(u, v)* is replaced by its inverted value,


*I(u, v)* ← 255 − *I(u, v),*


for all image coordinates *(u, v),* with *u = 0. . .M−1* and *v = 0. . .N−1*.


You should loop through each pixel and do the following:


1)    get the pixel value,


2)    compute the inverted value,


3)    save this final result into the current pixel location.


```matlab
hw1 = imread('hw1.png');
width = size(hw1,1);
length = size(hw1,2);
hw1_inverse = zeros(480,500);
for j = 1: length
    for i = 1:width
        hw1_inverse(i,j) = 255 - hw1(i,j);
    end
end
imagesc(hw1_inverse); colormap gray
```





**‘Secret\-Message’ filter**


This filter should also work on 8\-bit grayscale images.


You should loop through each pixel and do the following:


1)    get the pixel value,


2)    compute that value in mod two,


3)    multiply the result by 255, and


4)    save this final result into the current pixel location.


```matlab
hw1 = imread('hw1.png');
width = size(hw1,1);
length = size(hw1,2);
hw1_sm_filter = zeros(480,500);
for j = 1: length
    for i = 1:width
        hw1_sm_filter(i,j)=255*mod(hw1(i,j),2);
    end
end
imagesc(hw1_sm_filter); colormap gray
title('SM Filter 255 Multiplication')
```
